```python
gris = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gris,cmap="gray")
```

```python
batch = numpy.array([
    [-1,0,1],
    [-2,0,2],
    [-3,0,3]
])
```

```python
type(batch)
```

```python
def applyBatch(image,batch,x,y):
    s = 0
    for i in range(3):
        for j in range(3):
            s = s + image[x+i-1][y+j-1]*batch[i][j]
    return s
```

```python
resultat = numpy.zeros_like(gris)
print(gris.shape)
```

```python
for i in range(1,gris.shape[0]-1):
    for j in range(1,gris.shape[1]-1):
        resultat[i][j] = applyBatch(gris,batch,i,j)
```

```python
plt.imshow(gris,cmap="gray")
plt.show()
plt.imshow(resultat,cmap="gray")
plt.show()
```
