### impot

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
# from IPython import display
```

### Activate camera

```python
vc = cv2.VideoCapture(0)
```

### get frames

```python
if vc.isOpened():
    #get first frame
    is_capturing,frame = vc.read()
    # makes the blues image look like real colored
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame)
    plt.show()
else:
    is_capturing = False

while is_capturing:
    try:
        #Lookout for a keyboardInterrupt to stop the script
        is_capturing, frame = vc.read()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(frame)
        
        
        display.clear_output(wait = True)
        display.display(plt.gcf())
        # the pause time is =1 / framerate
        plt.pause(0.1)  
    except KeyboardInterrupt:
        vc.release()
```
